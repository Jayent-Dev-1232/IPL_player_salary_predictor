```python
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd

players_name = []
year = []
salary = []
teams = []
status = []

proxies_list = [
'http://45.67.214.86',
'http://104.16.133.224',
'http://31.43.179.101',
'http://104.17.155.21',
'http://104.17.186.228',
'http://104.19.51.216',
'http://31.43.179.22',
'http://172.67.183.130',
'http://104.19.84.74',
'http://104.18.112.32',
'http://164.38.155.221',
'http://160.153.0.150',
'http://154.83.2.166',
'http://170.114.46.179',
'http://154.197.75.2',
'http://141.101.121.51',
'http://45.131.7.48',
'http://172.67.144.227',
'http://185.162.231.218',
'http://104.27.71.47',
'http://104.16.0.167',
'http://45.159.219.131',
'http://123.25.234.143',
'http://45.12.30.254',
'http://195.85.23.5',
'http://104.21.18.107',
'http://45.131.7.189',
'http://154.92.9.102',
'http://185.162.228.241',
'http://172.67.181.16',
'http://23.227.60.7',
'http://104.17.217.76',
'http://104.17.30.106',
'http://172.67.161.206',
'http://104.21.63.235',
'http://5.10.246.213',
'http://104.18.13.72',
'http://185.148.104.199',
'http://5.10.246.101',
'http://23.227.38.205',
'http://104.17.110.189',
'http://194.152.44.206',
'http://154.83.2.146',
'http://104.19.93.125',
'http://104.18.19.210',
'http://185.221.160.20',
'http://104.21.192.75',
'http://5.10.244.195',
'http://104.16.139.188',
'http://104.17.62.167',
'http://185.162.230.103',
'http://94.73.239.124',
'http://104.25.237.201',
'http://45.131.5.57',
'http://104.24.222.17',
]

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/125.0.0.0 Safari/537.36"
}

proxy = random.choice(proxies_list)
proxy_dict = { "http": proxy}

url_of_teams = {
    "Chennai Super Kings": "https://iplt20stats.com/ipl/team/auction/chennai-super-kings",
    "Delhi Capitals": "https://iplt20stats.com/ipl/team/auction/delhi-capitals",
    "Gujarat Titans": "https://iplt20stats.com/ipl/team/auction/gujarat-titans",
    "Kolkata Knight Riders": "https://iplt20stats.com/ipl/team/auction/kolkata-knight-riders",
    "Lucknow Supergiants": "https://iplt20stats.com/ipl/team/auction/lucknow-supergiants",
    "Mumbai Indians": "https://iplt20stats.com/ipl/team/auction/mumbai-indians",
    "Punjab Kings": "https://iplt20stats.com/ipl/team/auction/punjab-kings",
    "Rajasthan Royals": "https://iplt20stats.com/ipl/team/auction/rajasthan-royals",
    "Royal Challengers Bangalore": "https://iplt20stats.com/ipl/team/auction/royal-challengers-bangalore",
    "Kings XI Punjab": "https://iplt20stats.com/ipl/team/auction/kings-xi-punjab",
    "Sunrisers Hyderabad": "https://iplt20stats.com/ipl/team/auction/sunrisers-hyderabad"
}
for team, url in url_of_teams.items():
  print(f"Scrapping data of {team}")
  r = requests.get(url, headers=headers, proxies=proxy_dict, timeout=10)
  soup = BeautifulSoup(r.text, 'html.parser')
  tables = soup.select('table.table-hover.mb-0')
  table = tables[1]
  rows = table.find_all('tr')

  for row in rows:
    cells = row.find_all('td')
    if len(cells) >=5:
      yr = int(cells[0].get_text(strip=True))
      if yr >= 2018 and yr <= 2025:
        year.append(yr)
        players_name.append(cells[1].get_text(strip=True))
        salary.append(cells[2].get_text(strip=True))
        status.append(cells[3].get_text(strip=True))
        teams.append(team)

data = {
    "Player Name": players_name,
    "Year": year,
    "Salary": salary,
    "Team": teams,
    "Retained/Bought/Unsold": status
}
df = pd.DataFrame(data)
df.to_csv("IPL_Auction_data_2018-2025.csv", index=False)
df.to_excel("IPL_Auction_data_2018-2025.xlsx", index=False)
```
